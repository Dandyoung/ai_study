```python
import numpy as np

def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / np.sum(exp_x)

# 케이스 1: 0으로 마스킹
scores_with_zero = np.array([2.0, 1.0, 0, 0])
probs_with_zero = softmax(scores_with_zero)

# 케이스 2: -inf로 마스킹
scores_with_inf = np.array([2.0, 1.0, -np.inf, -np.inf])
probs_with_inf = softmax(scores_with_inf)

print("0으로 마스킹한 경우:")
print("Scores:", scores_with_zero)
print("Softmax probabilities:", probs_with_zero)
print("Sum of probabilities:", np.sum(probs_with_zero))
print("\n마스킹된 위치의 확률:", probs_with_zero[2:], "\n")

print("-inf로 마스킹한 경우:")
print("Scores:", scores_with_inf)
print("Softmax probabilities:", probs_with_inf)
print("Sum of probabilities:", np.sum(probs_with_inf))
print("\n마스킹된 위치의 확률:", probs_with_inf[2:])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# x 값 범위 설정
x = np.linspace(-5, 5, 200)
y = np.exp(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', label='y = e^x')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('지수함수 (y = e^x)')
plt.legend()

# x축과 y축이 만나는 지점 표시
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)

# y축 범위 설정
plt.ylim(0, 20)

# 주요 포인트 표시
plt.plot(0, 1, 'ro', label='(0, 1)')
plt.text(0.2, 1.2, '(0, 1)')

plt.show()
```
