```python
import pandas as pd
import sqlite3

def csv_to_sqlite(csv_file_path, sqlite_db_path, table_name):
    # CSV 파일을 읽어 데이터프레임으로 변환
    df = pd.read_csv(csv_file_path)
    
    # SQLite 데이터베이스에 연결 (데이터베이스가 없으면 새로 생성됨)
    conn = sqlite3.connect(sqlite_db_path)
    
    # 데이터프레임을 SQLite 데이터베이스에 저장
    df.to_sql(table_name, conn, if_exists='replace', index=False)
    
    # 연결 닫기
    conn.close()

# 경로 및 테이블 이름 설정
csv_file_path = "/workspace/youngwoo/toyproject-datallm/modules/agent/gt_v1_head.csv"
sqlite_db_path = "/workspace/youngwoo/toyproject-datallm/modules/agent/gt_v1_head.db"
table_name = "data"

# 변환 실행
csv_to_sqlite(csv_file_path, sqlite_db_path, table_name)

print("CSV 파일이 SQLite 데이터베이스로 성공적으로 변환되었습니다.")
```

```python
import sqlite3

# 데이터베이스 파일 경로
db_path = '/workspace/youngwoo/toyproject-datallm/dataset/test_dataset.db'

# 쿼리
query = '''SELECT 
  AVG(Col1) AS AVG, 
    MAX(Col1) AS MAX, 
    MIN(Col1) AS MIN, 
    SUM(Col1) AS SUM, 
    COUNT(Col1) AS COUNT 
FROM test_table 
WHERE 
    (Col1 = 'G1 Generator Efficiency' AND 
    (Col1 = 'GT GEN#1 KV' AND 
    (Col1 = 'GT GEN#1 MVAR' AND 
    (Col1 = 'GT GEN#1 MEGAWATT TARGET') AND 
    (Col1 = 'GT GEN#1 Generator Power Factor Curve' AND 
    (Col1 = 'GT GEN#1 Generator Loss by Power Factor Curve' AND 
    (Col1 = 'GT GEN#1 Generator Loss by Power Factor Curve at Design Power Factor and Actual MW') AND 
    (Col1 = 'GT#1 Generator Loss by Power Factor Curve at Design Power Factor and Actual MW') AND 
    (Col1 = 'GT#1 Generator Exciter Loss by Curve' AND 
    (Col1 = 'GT#1 Generator Loss by Power Factor Curve at Design Power Factor and Actual MW') AND 
    (Col1 = 'GT#1 Generator Power Factor Actual') AND 
    (Col1 = 'GT#1 Generator Loss by Power Factor Curve at Design Power Factor and Actual MW') AND 
    (Col1 = 'GT#1 ppm NOx in Exhaust Gas') AND 
    (Col1 = 'GT#1 % O2 in Exhaust Flue Gas') AND 
    (Col1 = 'GT#1 Generator Auxirary (MW') AND 
    (Col1 = 'GT#1 Generator Gross Generation (MW') AND 
    (Col1 = 'GT#1 Generator Hydrogen Pressure') AND 
    (Col1 = 'GT#1 Compressor Discharge Pressure') AND 
    (Col1 = 'GT#1 Inlet Pressure Loss') AND 
    (Col1 = 'GT#1 Turbine Exhaust Draft Loss') AND 
    (Col1 = 'GT#1 NG WTR Pressure') AND 
    (Col1 = 'GT#1 Generator Cold Gas Temperature') AND 
    (Col1 = 'GT#1 Fuel Temperature') AND 
    (Col1 = 'S-TBQC-035 In air tot press tr') AND 
    (Col1 = 'GT#1 Fuel Temperature') AND 
    (Col1 = 'GT#1 Fuel Inlet Temperature') 
ORDER BY 
    (Col1) ASC
LIMIT 
    10
OFFSET 
    0
'''
# 데이터베이스와 연결하고 쿼리 실행
def run_query(db_path, query):
    # 데이터베이스에 연결
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()

    # 쿼리 실행
    cursor.execute(query)
    
    # 결과 가져오기
    results = cursor.fetchall()
    
    # 열 이름 가져오기
    columns = [description[0] for description in cursor.description]

    # 연결 닫기
    conn.close()
    
    # 결과 출력
    return columns, results

# 메인 함수
def main():
    columns, results = run_query(db_path, query)
    print("Columns:", columns)
    for row in results:
        print(row)

if __name__ == "__main__":
    main()
```

```python
# 각 테이블의 열 타입을 얻는 함수
def get_table_schema(db_path, table_name):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()
    cursor.execute(f"PRAGMA table_info({table_name});")
    schema = cursor.fetchall()
    conn.close()
    return schema

get_table_schema('/workspace/youngwoo/toyproject-datallm/dataset/test_dataset.db', 'test_table')
```
